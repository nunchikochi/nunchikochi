```python
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
iris
```

```python
iris.describe().T
```

```python
sns.displot(data=iris,x='sepal_length',hue='species',kind='kde') #hue='분류_기준', kind='kde' 내재된 확률분포를 그려줌
```
